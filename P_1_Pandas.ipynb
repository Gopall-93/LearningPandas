```python
import pandas as pd
```

Imported a dataset from kaggle and took a path of it and stored it in a path variable

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("pratsharma7/student-quiz-marks-dataset")

print("Path to dataset files:", path)
```

read the data set using pd.read_csv() function from pandas

```python
dataset = pd.read_csv(path+"/marks_final.csv")
```

data.head() gives us the table

```python
dataset.head()
```

To get the how many elements does the dataframe have

```python
dataset.info()
```

Shape of the dataset

```python
dataset.shape
```

**loc[index,column]** is used to get the row with the specidfied index in our case the index is the SNO. containing number 1,2,3...so on but the index can be string or number or any other data type

```python
dataset.loc[0,"Q1 (5)"]
```

What if we want the **MAXIMUM value** in particular column we can use **.max()** to get it

```python
dataset['Out of 30'].max()
```

```python
dataset.max()
```

To get the statistic of the whole dataset we can use **.describe()** function

```python
dataset.describe()
```

```python
dataset["Out of 30"].describe()
```

If we want to locate multiple indexs or multiple columns we can give **.loc[ [ ],[ ] ]** an array of index to search for

```python
dataset.loc[[1,2,50,100],["Q1 (5)","Top 9 (out of 45)"]]
```

```python
dataset.describe().loc["mean"]
```

```python
dataset["Out of 30"] >= 25
```
